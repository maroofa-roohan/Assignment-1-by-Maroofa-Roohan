```python



#differnt types of operators in python


#1Arithmetic Operators
##Used for basic mathematical operations:
#+ (Addition):
#- (Subtraction)
#* (Multiplication):
#/ (Division):
#% (Modulus)
#** (Exponentiation)
#// (Floor Division)

#2. Comparison Operators
#Used to compare values:
#== (Equal to):
#!= (Not equal to)
#> (Greater than):
#< (Less than):
#>= (Greater than or equal to):
#<= (Less than or equal to):

#3. Logical Operators
#Used to perform logical operations:
#and (Logical AND):
#or (Logical OR):
#not (Logical NOT):

#4. Assignment Operators
#Used to assign values to variables:
#= (Assignment):
#+= (Add and assign):
#-= (Subtract and assign):
#v*= (Multiply and assign):
#/= (Divide and assign):
#%= (Modulus and assign):
#**= (Exponentiation and assign)
#//= (Floor division and assign)

#5. Bitwise Operators
#Used for bit-level operations:
#& (Bitwise AND):
#| (Bitwise OR)
#^ (Bitwise XOR):
#~ (Bitwise NOT):
#<< (Left shift):
#>> (Right shift):
#
#6. Identity Operators
#Used to compare objects by their identity:

#7. Membership Operators
#Used to check membership in a sequence:


#Arithmetic Operators
# Addition
a = 12 + 7
print(a)  # Output: 19

# Subtraction
a = 15 - 5
print(a)  # Output: 10

# Multiplication
a = 8 * 6
print(a)  # Output: 48

# Division
a = 20 / 4
print(a)  # Output: 5.0

# Modulus
a = 14 % 5
print(a)  # Output: 4

# Exponentiation
a = 2 ** 4
print(a)  # Output: 16

# Floor Division
a = 17 // 3
print(a)  # Output: 5

#Comparison Operators

# Equal to
k = (8 == 8)
print(k)  # Output: True

# Not equal to
k = (8 != 8)
print(k)  # Output: False

# Greater than
k = (8 > 5)
print(k)  # Output: True

# Less than
k = (8 < 10)
print(k)  # Output: True

# Greater than or equal to
k = (8 >= 8)
print(k)  # Output: True

# Less than or equal to
k = (8 <= 7)
print(k)  # Output: False

#Logical Operators

# Logical AND
m = (10 > 5 and 3 < 7)
print(m)  # Output: True

# Logical OR
m = (10 > 5 or 3 > 7)
print(m)  # Output: True

# Logical NOT
m = not (10 > 5)
print(m)  # Output: False

#Assignment Operators
# Assignment
l = 10

# Add and assign
l += 5
print(l)  # Output: 15

# Subtract and assign
l -= 3
print(l)  # Output: 12

# Multiply and assign
l *= 2
print(l)  # Output: 24

# Divide and assign
l /= 6
print(l)  # Output: 4.0

# Modulus and assign
l %= 3
print(l)  # Output: 1.0

# Exponentiation and assign
l **= 2
print(l)  # Output: 1.0

# Floor division and assign
l //= 1
print(l)  # Output: 1.0


#Bitwise Operators

# Bitwise AND
d = 6 & 3
print(d)  # Output: 2

# Bitwise OR
d = 6 | 3
print(d)  # Output: 7

# Bitwise XOR
d = 6 ^ 3
print(d)  # Output: 5

# Bitwise NOT
d = ~6
print(d)  # Output: -7

# Left shift
d = 6 << 2
print(d)  # Output: 24

# Right shift
d = 6 >> 1
print(d)  # Output: 3


#Identity Operators
# Identity comparison
list_a = [1, 2, 3]
list_b = list_a
r = (list_a is list_b)
print(r)  # Output: True

# Identity comparison negation
list_c = [1, 2, 3]
r = (list_a is not list_c)
print(r)  # Output: True


#Membership Operators

# Membership check
c = ('e' in 'hello')
print(c)  # Output: True

# Membership check negation
c = ('x' not in 'hello')
print(c)  # Output: True
```

```python
#typecasting
#Typecasting in programming refers to the conversion of a variable from one data type to another.The two types as follows
#Implicit vs. Explicit Typecasting:

#Implicit Typecasting: Also known as automatic type conversion, it happens automatically by
#the compiler or interpreter when you perform operations between different data types. For example, adding an
#integer and a float will automatically convert the integer to a float.


#Explicit Typecasting: Requires the programmer
#to manually convert the data type using typecasting functions or methods. For example,
#converting a string to an integer.




#example

#Implicit Typecasting:

a = 5        # integer
b = 2.0      # float
c = a + b    # implicit conversion of integer to float
print(c)     # Output: 7.0


#Explicit Typecasting:

# Convert integer to float
num = 10
float_num = float(num)
print(float_num)  # Output: 10.0

# Convert float to integer
num = 10.7
int_num = int(num)
print(int_num)    # Output: 10

# Convert string to integer
str_num = "123"  #ALAPHABETS not possible as string to convert in integer
int_num = int(str_num)
print(int_num)    # Output: 123

# Convert integer to string
num = 456
str_num = str(num)
print(str_num)    # Output: "456"



```

```python
#Loops are control structures used in programming to repeat a block of code multiple times based on certain conditions. Python supports several types of loops,
#each with specific use cases
# the different types of loops in Python:
#for Loop
#while Loop


#example for loop
for i in range(5):
    print(i)

#while loop
count = 1
while count <= 5:
    print(count)
    count += 1

#The break and continue statements are used to control the execution flow within loops









#The break and continue statements are used to control the execution
#flow within loops

#####################################

#break Statement
#Immediately exits the loop, regardless of the loop's condition.
#Stops the loop entirely and transfers control to the statement immediately following the loop.
#The loop does not continue to the next iteration if break is executed.
#It is mainly used to exit a loop when a certain condition is met, often used to terminate the loop early.
#Example:


for i in range(10):
    if i == 5:
        break
    print(i)

# continue Statement
#Skips the rest of the code inside the loop for the current iteration and moves to the next iteration.
#The loop continues to run, but any code following the continue statement in the current iteration is skipped.
#The loop condition is re-evaluated, and if it's still True, the next iteration starts.
#Use Case: To skip certain iterations of a loop based on a condition, often used to avoid executing specific code under certain circumstances.
#Example:

for i in range(10):
    if i == 5:
        continue
    print(i)

#Both break and continue are powerful tools for controlling loop execution
#and managing how iterations are handled
```
